Interprétation :
Cette ligne installe une bibliothèque appelée rake-nltk, qui permet d’extraire automatiquement des mots-clés à partir d’un texte, en s’appuyant sur des techniques de Text Mining et de traitement automatique du langage (NLP).

```python
text1 ="""During travelling you may experience the craziest, most exciting things, that will eventually become great stories that you can tell others. When you grow old and look back at your life and all your travel experiences, you will realise how much you have done in your life and your life was not in vain. It can provide you with happiness and satisfaction for the rest of your life."""
from rake_nltk  import Rake
 # Uses stopwords for english from NLTK, and all puntuation characters by default
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
r = Rake()
# Extraction given the text.
r.extract_keywords_from_text(text1)
print(r)
 # Extraction given the list of strings where each string is a sentence.
#r.extract_keywords_from_sentences(<list of sentences>)
# To get keyword phrases ranked highest to lowest.
r.get_ranked_phrases() [0:5]
 # To get keyword phrases ranked highest to lowest with scores.
r.get_ranked_phrases_with_scores() [0:5]


```

Interprétation : 1- Importation et téléchargement des ressources
Le code importe la bibliothèque rake_nltk et nltk, puis télécharge :

les stopwords pour ignorer les mots vides comme the, and, is,

le tokenizer punkt_tab pour découper le texte en phrases et mots.

2- Création de l’objet RAKE
L’objet r = Rake() permet d’appliquer l’algorithme RAKE sur le texte pour extraire automatiquement les mots-clés ou expressions significatives.

3- Extraction des mots-clés
r.extract_keywords_from_text(text1) analyse le texte, supprime les stopwords et la ponctuation, puis identifie les expressions les plus importantes selon leur fréquence et cooccurrence.

3- Résultats

r.get_ranked_phrases_with_scores() retourne les expressions clés avec leur score d’importance.

Interprétation des scores obtenus : Plus le score est élevé, plus l’expression est représentative du texte.

Ici, l’expression “eventually become great stories” a le score le plus élevé → c’est le thème central du texte.

Les autres expressions comme “travel experiences” ou “look back” représentent des thèmes secondaires mais importants.

Interprétation : Cette ligne sert simplement à installer YAKE
C’est un algorithme non supervisé pour l’extraction automatique de mots-clés.
Différence avec RAKE : YAKE ne dépend pas des stopwords NLTK et peut être plus efficace sur des textes courts.

```python
import yake
kw_extractor = yake.KeywordExtractor()
text = """During travelling you may experience the craziest, most exciting things, that will eventually become great stories that you can tell others. When you grow old and look back at your life and all your travel experiences, you will realise how much you have done in your life and your life was not in vain. It can provide you with happiness and satisfaction for the rest of your life."""
language = "en"
max_ngram_size = 3
deduplication_threshold = 0.9
numOfKeywords = 5
custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
keywords = custom_kw_extractor.extract_keywords(text)
for kw in keywords:
    print(kw)
```

Interprétation : Chaque élément est un tuple (mot-clé, score).

Le premier élément → l’expression clé.

Le second → le score d’importance, de type np.float64.

Règle de YAKE : plus le score est petit, plus le mot-clé est important.

"exciting things" avec 0.0219 est plus représentatif que "experience the craziest" avec 0.0815.

```python
#How to use it on Python
import yake
text = "During travelling you may experience the craziest, most exciting things, that will eventually become great stories that you can tell others. When you grow old and look back at your life and all your travel experiences, you will realise how much you have done in your life and your life was not in vain. It can provide you with happiness and satisfaction for the rest of your life."
#assuming default parameters
kw_extractor = yake.KeywordExtractor()
keywords = kw_extractor.extract_keywords(text)
for kw in keywords:
	print(kw)
#specifying parameters
language = "en"
#max_ngram_size = 3
deduplication_thresold = 0.1
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 5
custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
keywords = custom_kw_extractor.extract_keywords(text)
for kw in keywords:
    print(kw)
```

Interprétation : Cette partie montre l’utilisation de YAKE avec paramètres par défaut et paramètres personnalisés.
Les mots-clés les plus importants sont "exciting things", "great stories", "eventually become great" et "life", reflétant les thèmes principaux du texte : expériences de voyage, transformation en histoires et sens de la vie.
